```python
# importação das bibliotecas

import pandas as pd
from pandas_datareader import data
import yfinance as yf
     
```

```python
# criação do dataframe

acoes = ['AMER3.SA', 'ODPV3.SA', 'CCRO3.SA', 'PETR4.SA', 'BBAS3.SA', 'JBSS3.SA','ABEV3.SA','SQIA3.SA', 
         'RENT3.SA','LREN3.SA']


acoes_df = pd.DataFrame()
for acao in acoes:
  #acoes_df[acao] = data.DataReader(acao, data_source='yahoo', start='2015-01-01')['Close']
  acoes_df[acao] = yf.download(acao, start='2015-01-02')['Close']

acoes_df  
```

```python
#getting only the data we want
acoes_df = acoes_df.iloc[:, :10]

acoes_df
```

```python
#Definição das datas

start = "2015-01-02 "
training_end = "2023-01-05"
treatment_start = "2023-01-12"
end = "2023-01-27"
end_stock = "2023-01-30"
```

```python
acoes_df.plot()
```

```python
# dataframe com apenas periodo de tratamento

acoes_df_cor = acoes_df[acoes_df.index <= treatment_start]

# matriz de correlação 

acoes_df_cor.corr()
pd.set_option('display.max_columns', None)

```

```python
acoes_df_cor.corr()
```

```python
# seleção das acoes com maior correlaçao com AMER3

final_series = acoes_df[["AMER3.SA", "SQIA3.SA"	,"LREN3.SA", "RENT3.SA","ABEV3.SA"]]

```

```python
final_series.plot()
```

```python
# criacao do periodo pre e pos 

pre_period = [start, training_end]
post_period = [treatment_start, end]
```

```python
pip install pycausalimpact
```

```python
#import library


from causalimpact import CausalImpact
impact = CausalImpact(data = final_series,
                      pre_period = pre_period,
                      post_period = post_period,
                      nseasons = [{'period': 52}],
                      priod_level_sd = None)
```

```python
# plotando resultados graficos

impact.plot()

```

```python
# relatório 

print(impact.summary('report'))
```
